```python
import requests
from skimage import io
import matplotlib.pyplot as plt

url = "https://dog.ceo/api/breeds/image/random"
res = requests.get(url)
data = res.json()
image_url = data.get('message')
image = io.imread(image_url)
plt.axis('off')
plt.imshow(image)
```
